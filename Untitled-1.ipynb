```python
import os
import pandas as pd

# Define the directory where you want to check for the folders
directory_to_check = "/home/vault/empkins/tpD/D02/processed_data/SonyCam_processed_data"

# Read the CSV file
df = pd.read_csv("/home/vault/empkins/tpD/D02/processed_data/sonycam_videos_dir2.csv")

# Define a function to check if a folder exists
def check_folder_exists(patient_id):
    folder_path = os.path.join(directory_to_check, str(patient_id))
    return os.path.exists(folder_path)

# Add a new column to the dataframe indicating whether the folder exists or not
df['Processing Status'] = df['Patient_ID'].apply(check_folder_exists)
```

```python
# Save the updated dataframe back to the CSV file
df.to_csv("/home/vault/empkins/tpD/D02/Students/Aditya/sony_status.csv", index=False)
```
